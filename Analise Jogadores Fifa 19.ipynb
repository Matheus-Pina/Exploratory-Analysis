No dataframe, existem três colunas que trazem valores financeiros. Irei verificar seu tipo e fazer as necessárias alterações

```python
df[['Wage', 'Value', 'Release Clause']].info()
```

Como essas colunas são do tipo Object, preciso transformá-las para o tipo Float

```python
# Retirando os cifrões

df['Wage'] = df['Wage'].str.replace('€', '')
df['Value'] = df['Value'].str.replace('€', '')
df['Release Clause'] = df['Release Clause'].str.replace('€', '')
```

```python
df[['Wage', 'Value', 'Release Clause']].head()
```

Agora, irei multiplicar o número por 1.000 (substituindo o K) ou 1.00.000 (substituindo o M)

```python
df.Value = (df.Value.replace(r'[KM]', '', regex=True).astype(float) * \
          df.Value.str.extract(r'[\d\.]+([KM]+)', expand=False)\
          .fillna(1).replace(['K','M'], [10**3, 10**6]).astype(int))
```

```python
df.Wage = (df.Wage.replace(r'[KM]', '', regex=True).astype(float) * \
          df.Wage.str.extract(r'[\d\.]+([KM]+)', expand=False)\
          .fillna(1).replace(['K','M'], [10**3, 10**6]).astype(int))
```

```python
df['Release Clause'] = (df['Release Clause'].replace(r'[KM]', '', regex=True).astype(float) * \
          df['Release Clause'].str.extract(r'[\d\.]+([KM]+)', expand=False)\
          .fillna(1).replace(['K','M'], [10**3, 10**6]).astype(int))
```

```python
df[['Wage', 'Value', 'Release Clause']].info()
```

# Exploratory Data Analysis

* Irei explorar as informações que o conjunto de dados têm, o que me permitirá tirar conclusões acerca dos jogadores e times do Fifa 2019

```python
# Criando um novo dataframe que conterá a média salarial dos times
# Ao fim, analiso como os dados se distribuem

club_wage_mean = df[['Club', 'Wage']].groupby('Club').mean()
club_wage_mean = club_wage_mean.sort_values(by = 'Wage', ascending = False)
club_wage_mean.describe()
```

```python
plt.figure(figsize = (10,10))
club_wage_mean.boxplot()
plt.title('Boxplot da Média Salarial', fontsize = 20, pad = 20)
plt.show()
```

Usando um BoxPlot podemos verificar a presença de alguns outliers. Isso explica a enorme variação entre o terceiro quartil e o valor máximo apresentados acima. Além disso, podemos perceber que 50% dos clubes possuem uma média salarial entre 1000 e 3538.Enquanto os outros 50% variam entre 3538 e 152030. O que demonstra uma enorme desigualdade entre os times.

### Agora irei filtrar para indenticar quais são os clubes que são mostrados como outliers no gráfico

```python
plt.figure(figsize = (10,10))
club_wage_mean[club_wage_mean['Wage'] >= 20000].boxplot()
plt.title('Boxplot da Média Salarial\ndos Times com Média acima de 20.000', fontsize = 20, pad = 20)
plt.show()
```

Ainda assim, permancem muitas diferenças, gerando alguns outlier. Assim, irei buscar os valores acima de 30000.

```python
# Estes são os 5 clubes com maiores médias salariais

high_wage = club_wage_mean[club_wage_mean['Wage'] >= 30000]
high_wage.head()
```

```python
high_wage.plot(kind = 'bar', figsize = (15,10), color = 'steelblue')
plt.title('Média Salarial por Time', fontsize = 20, pad = 20)
plt.show()
```

A partir do gráfico acima, percebemos a baixa ou moderada relação entre as variáveis, com exceção das seguintes:

```python
ax = sns.jointplot(x = 'Release Clause', 
           y = 'Wage',
          data=df,
             kind = 'reg', height= 7)

ax.fig.suptitle("Correlação entre Wage e Release Clause\n(com a linha da regressão)",
                fontsize = 18,
               y = 1.05)

plt.show()
```

```python
ax = sns.jointplot(x = 'Value', 
           y = 'Release Clause',
          data=df,
             kind = 'reg', height= 7)

ax.fig.suptitle("Correlação entre Release Clause e Value\n(com a linha da regressão)",
                fontsize = 18,
               y = 1.05)

plt.show()
```

A partir dos dois gráficos acima, podemos perceber relação entre as variáveis. No primeiro caso, a correlação (r = 0.86) é significante, porém, quando comparada à correlação quase perfeita entre Release Clause e Value, é notório um comportamento distinto dos valores, estando quase todos sobre a linha da regressão. Assim, quanto maior for o valor de um jogador, naturalmene, maior será o valor da rescisão contratual. O mesmo também ocorre entre rescisão contratual e salário, não obstante, em menor intensidade.
